1. transformers[sentencepiece]
Transformers is a widely-used NLP library developed by Hugging Face. It provides a variety of pre-trained models (such as BERT, GPT, T5, etc.) for tasks like text classification, translation, summarization, and more.
[sentencepiece]: The sentencepiece is an optional dependency for tokenization, particularly useful for models like T5, MarianMT, and others that use SentencePiece-based tokenization. It helps with tokenizing text into subword units in a language-agnostic way.

2. datasets
This is another library by Hugging Face, designed for easy access to NLP datasets. It provides a collection of ready-to-use datasets like SQuAD, IMDB, and many more, making it easy to download, preprocess, and use datasets directly in machine learning workflows.
It’s optimized for large-scale datasets and provides efficient features for streaming data, caching, and batch processing.

3. sacrebleu
This is a tool for evaluating machine translation models. It calculates the BLEU (Bilingual Evaluation Understudy) score, which is a metric for comparing machine-translated text against one or more reference translations.
SacreBLEU ensures that BLEU scores are calculated consistently across different models and research papers by standardizing the metric.

4. rouge_score
This package provides the implementation of ROUGE (Recall-Oriented Understudy for Gisting Evaluation), another common metric used to evaluate text generation tasks such as summarization.
ROUGE evaluates the quality of a machine-generated text by comparing it to a reference, measuring the overlap of n-grams, word sequences, and word pairs.

5. py7zr
This is a Python library that allows reading and writing of 7-zip files (.7z). It’s used to handle compressed files in this format, which is often used to distribute large datasets or models in a compressed manner.

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"

device
```

```python

model_ckpt = "google/pegasus-cnn_dailymail"   #  used model

tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)

```

```python
#load the data

dataset_samsum = load_from_disk('samsum_dataset')
dataset_samsum
```

```python
def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(example_batch['dialogue'] , max_length = 1024, truncation = True )

    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['summary'], max_length = 128, truncation = True )

    return {
        'input_ids' : input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched = True)
```

```python
dataset_samsum_pt["train"]
```

```python
# Training

from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus) #creating batches
```

```python



from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='pegasus-samsum', num_train_epochs=1, warmup_steps=500,
    per_device_train_batch_size=1, per_device_eval_batch_size=1,
    weight_decay=0.01, logging_steps=10,
    evaluation_strategy='steps', eval_steps=500, save_steps=1e6,
    gradient_accumulation_steps=16
)
```

```python
trainer = Trainer(model=model_pegasus, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["test"],
                  eval_dataset=dataset_samsum_pt["validation"])
```

```python
trainer.train()
```

```python
# Evaluation

def generate_batch_sized_chunks(list_of_elements, batch_size):
    """split the dataset into smaller batches that we can process simultaneously
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]



def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                               batch_size=16, device=device,
                               column_text="article",
                               column_summary="highlights"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)):

        inputs = tokenizer(article_batch, max_length=1024,  truncation=True,
                        padding="max_length", return_tensors="pt")

        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                         attention_mask=inputs["attention_mask"].to(device),
                         length_penalty=0.8, num_beams=8, max_length=128)
        ''' parameter for length penalty ensures that the model does not generate sequences that are too long. '''

        # Finally, we decode the generated texts,
        # replace the  token, and add the decoded texts with the references to the metric.
        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                clean_up_tokenization_spaces=True)
               for s in summaries]

        decoded_summaries = [d.replace("", " ") for d in decoded_summaries]


        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    #  Finally compute and return the ROUGE scores.
    score = metric.compute()
    return score
```

```python
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]  #text summarization metrics
rouge_metric = evaluate.load('rouge')
```

```python
score = calculate_metric_on_test_ds(
    dataset_samsum['test'][0:10], rouge_metric, trainer.model, tokenizer, batch_size = 2, column_text = 'dialogue', column_summary= 'summary'
)

#print(score)

#rouge_dict = dict((rn, score[rn].mid.fmeasure ) for rn in rouge_names )

#pd.DataFrame(rouge_dict, index = [f'pegasus'] )
# Directly use the score dictionary as it already contains the final values
rouge_dict = dict((rn, score[rn]) for rn in rouge_names)

# Create a DataFrame from the dictionary
rouge_df = pd.DataFrame(rouge_dict, index=[f'pegasus'])

# Display the DataFrame
print(rouge_df)
```
